# Ridge

```python
score = []
cate=[]
for e_name in data1['Emission Name'].unique():
    for t_fuel in data1['Test Fuel'].unique():
        
        temp=data1[(data1['Emission Name']==e_name)&\
                  (data1['Test Fuel']==t_fuel)]
        if len(temp) < 10:
            continue
        temp=pd.concat([temp,pd.get_dummies(temp['Test Procedure'])], axis=1)
        temp.drop(columns=['Model Year', 'Certificate Manufacturer Name', 'Certified Test Group',\
                  'Certified Evaporative Family','Represented Test Vehicle Make','Represented Test Vehicle Model',\
                   'Vehicle ID','Vehicle Configuration Number', 'Gross Vehicle Weight Rating (lbs.)',\
                  'Equivalent Test Weight (lbs.)', 'Test Drive','Transmission Type Description',\
                  'Transmission type, if other','Transmission Lockup Yes/No?', 'Creeper Gear Yes/No?',\
                  'Vehicle Fuel Category Description','Hybrid Yes/No?','Test Procedure Description',\
                  'Test Fuel Description', 'Certification/In-Use Code', 'Vehicle Class',\
                   'Vehicle Class Description', 'Certification Region','Certification Region Description', \
                  'Additive Deterioration Factor', 'Multiplicative Deterioration Factor',\
                   'Using NMOG/NMHC Ratio?', 'Ratio of NMOG to NMHC','Test Fuel',\
                   'Upward Diesel Adjustment Factor', 'Downward Diesel Adjustment Factor',\
                   'Reactivity Factor (RAF)', 'Carline Models Covered','Test Drive Description',\
                  'Vehicle Fuel Category','Test Originator','Test Number','Emission Standard Level Code',\
                    'Emission Standard Level Description','Emission Name',\
                  'Target Coefficient A (lbf)','Target Coefficient B (lbf/mph)',\
                'Test Procedure','Target Coefficient C (lbf/mph**2)','Transmission Type'],\
                  inplace=True)
        temp.drop_duplicates(inplace=True)
        temp.dropna(inplace=True)
        y = temp['Rounded Emission Result (g/mi)']
        temp.drop(columns=['Rounded Emission Result (g/mi)'],inplace=True)
        if len(temp) > 100:
            X_train, X_test, y_train, y_test = train_test_split(temp, y, random_state=0)
            rid = linear_model.Ridge(alpha=0.05).fit(X_train, y_train)
            score.append(rid.score(X_test, y_test))
        else:
            rid = linear_model.Ridge(alpha=0.05).fit(temp, y)
            score.append(rid.score(temp, y))
        cate.append((e_name,t_fuel))
        
        
binwidth = 0.06
plt.hist(score, bins=np.arange(min(score), max(score) + binwidth, binwidth))
x=np.median(np.arange(min(score), max(score) + binwidth, binwidth))
y=max(np.histogram(score)[0])/5
plt.text(x*2,y,f'Average score is: {np.mean(score):.2f}')
plt.title('Histogram of R^2 score for ridge regression with alpha = 0.05')
plt.show()
```

```python
from sklearn.linear_model import LassoCV
score = []
cate=[]
top_1 = []
top_2 = []
top_3 = []
for e_name in data1['Emission Name'].unique():
    for t_fuel in data1['Test Fuel'].unique():
        
        temp=data1[(data1['Emission Name']==e_name)&\
                  (data1['Test Fuel']==t_fuel)]
        if len(temp) < 10:
            continue
        temp=pd.concat([temp,pd.get_dummies(temp['Test Procedure'])], axis=1)
        temp.drop(columns=['Model Year', 'Certificate Manufacturer Name', 'Certified Test Group',\
                  'Certified Evaporative Family','Represented Test Vehicle Make','Represented Test Vehicle Model',\
                   'Vehicle ID','Vehicle Configuration Number', 'Gross Vehicle Weight Rating (lbs.)',\
                  'Equivalent Test Weight (lbs.)', 'Test Drive','Transmission Type Description',\
                  'Transmission type, if other','Transmission Lockup Yes/No?', 'Creeper Gear Yes/No?',\
                  'Vehicle Fuel Category Description','Hybrid Yes/No?','Test Procedure Description',\
                  'Test Fuel Description', 'Certification/In-Use Code', 'Vehicle Class',\
                   'Vehicle Class Description', 'Certification Region','Certification Region Description', \
                  'Additive Deterioration Factor', 'Multiplicative Deterioration Factor',\
                   'Using NMOG/NMHC Ratio?', 'Ratio of NMOG to NMHC','Test Fuel',\
                   'Upward Diesel Adjustment Factor', 'Downward Diesel Adjustment Factor',\
                   'Reactivity Factor (RAF)', 'Carline Models Covered','Test Drive Description',\
                  'Vehicle Fuel Category','Test Originator','Test Number','Emission Standard Level Code',\
                    'Emission Standard Level Description','Emission Name',\
                  'Target Coefficient A (lbf)','Target Coefficient B (lbf/mph)',\
                'Test Procedure','Target Coefficient C (lbf/mph**2)','Transmission Type'],\
                  inplace=True)
        temp.drop_duplicates(inplace=True)
        temp.dropna(inplace=True)
        y = temp['Rounded Emission Result (g/mi)']
        temp.drop(columns=['Rounded Emission Result (g/mi)'],inplace=True)
        if len(temp) < 5:
            continue
        if len(temp) > 100:
            X_train, X_test, y_train, y_test = train_test_split(temp, y, random_state=0)
            clf =  LassoCV().fit(X_train, y_train)
#             score.append(rid.score(X_test, y_test))
        else:
            clf =  LassoCV().fit(temp, y)
#             score.append(rid.score(temp, y))
            
        importance = np.abs(clf.coef_)
        idx_third = importance.argsort()[-4]
        threshold = importance[idx_third] + 0.01
        idx_features = (-importance).argsort()[:3]
        feature_names=temp.columns.values
        name_features = np.array(feature_names)[idx_features]
        top_1.append(name_features[0])
        top_2.append(name_features[1])
        top_3.append(name_features[2])
        cate.append((e_name,t_fuel))
        
        
# binwidth = 0.06
# plt.hist(score, bins=np.arange(min(score), max(score) + binwidth, binwidth))
# x=np.median(np.arange(min(score), max(score) + binwidth, binwidth))
# y=max(np.histogram(score)[0])/2
# plt.text(x,y,f'Average score is: {np.mean(score):.2f}')
# plt.title('Histogram of R^2 score for Lasso regression with alpha = 0.01')
# plt.show()
```
